 How do player ratings vary across age groups, and what role does speed play in influencing overall performance?
 

# Dataset Documentation:

URL: https://www.kaggle.com/datasets/antoinekrajnc/soccer-players-statistics?resource=download

Dataset Description:  This dataset contains detailed information on thousands of professional soccer players from various nationalities and clubs. It includes personal attributes such as name, age, nationality, and preferred foot, as well as club and national team affiliations. Physical metrics like height and weight are included alongside a wide range of skill ratings — from ball control and dribbling to defensive abilities and goalkeeping. These attributes offer a comprehensive view of each player's strengths and weaknesses, making the dataset ideal for analyzing performance trends, scouting potential, and exploring how age, physical traits, and skill ratings relate to overall player effectiveness.

Time Period: 2018

Why this dataset: I chose this dataset because soccer is my favorite sport to watch and I wanted to learn more about individual players skill-based attributes. Understanding the ratings and how speed influences peformances.


```python
import pandas as pd

df = pd.read_csv("s3://ntia340/SoccerPlayersData.csv")
df.head(15)
```

# Data Cleaning

```python
df.info()
```

```python
missing_summary = df.isnull().sum()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['Name', 'Age', 'Nationality'])
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
print(df['Age'].value_counts().sort_index())
```

```python
f = df[(df['Age'] >= 15) & (df['Age'] <= 45)]
```

```python
print("\nCleaned dataset shape:", df.shape)
```

```python
df.reset_index(drop=True, inplace=True)
```

## New Variable: Age Group

```python
# Create age group bins
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[15, 20, 25, 30, 35, 40, 45],
    labels=['15–20', '21–25', '26–30', '31–35', '36–40', '41–45']
)

# Preview the new column
df[['Name', 'Age', 'AgeGroup']].head()
```

```python
print(df.columns.tolist())
```

## Summary Table: Average Rating by Age Group

```python
rating_summary = df.groupby('AgeGroup')['Rating'].mean().reset_index()

# Rename columns for clarity
rating_summary.columns = ['Age Group', 'Average Rating']

# Display the summary table
print(rating_summary)
```

## Graph 1 (Average Player Rating by Age Group)

```python
import matplotlib.pyplot as plt

# Bar chart of average rating by age group
rating_summary = df.groupby('AgeGroup')['Rating'].mean().reset_index()

plt.figure(figsize=(8, 5))
plt.bar(rating_summary['AgeGroup'], rating_summary['Rating'], color='forestgreen')
plt.title('Average Player Rating by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Rating')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

## Graph 2 ( Rating vs Speed)

```python
# Scatter plot of Rating vs Speed
plt.figure(figsize=(8, 5))
plt.scatter(df['Speed'], df['Rating'], alpha=0.6, color='skyblue')
plt.title('Player Rating vs Speed')
plt.xlabel('Speed')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```
